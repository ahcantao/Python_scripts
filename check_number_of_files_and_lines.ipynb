## This code is used to go through 9350 folders and read (at least)112,200 .txt files
#### I am using to check if the number of files (12) AND if the number of lines (41) on each file is correct

```python
#using glob to get each file path
import glob
import datetime

def check_output_files(base_path,noises,datasets,output_file_name):
    #going through each dataset/kfold/noise/score folder
    for dataset in datasets:
        print(f"working with dataset: {dataset}")
        log_file = open(output_file_name,"a")
        for kfold in range(1,11):
            for noise in noises:
                file_path = base_path+f"/{dataset}/{kfold:0>3}/{noise}/score/*"
                files = glob.glob(file_path)
                total_score_files = int(len(files))
                #checking the number of files on this folder
                if total_score_files < 12 or total_score_files > 12:
                    log_file.write(f"{dataset}-{kfold:0>3}-{noise}-{total_score_files}_files_of_12_ERROR\n")
                for file in files:
                    line_counter = 0
                    #counting the number of lines on this file
                    for line in open(file, "r"):
                        line_counter = line_counter + 1
                    if line_counter == 41:
                        log_file.write(f"{dataset}-{kfold:0>3}-{noise}-41_lines_OK\n")
                    else:
                        log_file.write(f"{dataset}-{kfold:0>3}-{noise}-{line_counter}_lines_ERROR\n")
        log_file.close()
    print("finished")
```

```python
start = datetime.datetime.now()
output_file_name = "check_number_of_files_and_lines_regression.txt"

#path to the datasets folders
base_path = "./regression"

#directories to be explored
noises = ['001','002','004','008','016','032','064','128','256','512','1024']

#getting datasets folders names into a list
datasets_path = glob.glob(base_path+"/*")
datasets = []
for dataset in datasets_path:
    datasets.append(dataset.split("\\")[1])
print("Number of datasets to be checked: ", len(noises))

#executing the verification
check_output_files(base_path,noises,datasets,output_file_name)

print("Wall time: ", datetime.datetime.now() - start)
```

```python
start = datetime.datetime.now()
output_file_name = "check_number_of_files_and_lines_classification.txt"

#path to the datasets folders
base_path = "./classification"

#directories to be explored
noises = ['001','002','004','008','016','032','064','128','256','512','1024']

#getting datasets folders names into a list
datasets_path = glob.glob(base_path+"/*")
datasets = []
for dataset in datasets_path:
    datasets.append(dataset.split("\\")[1])
print("Number of datasets to be checked: ", len(noises))

#executing the verification
check_output_files(base_path,noises,datasets,output_file_name)

print("Wall time: ", datetime.datetime.now() - start)
```
